### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
'''
File I/O的內容會一併將形成資料結構的標籤讀出，並且全部以字串的方式呈現，
而xmltodict所讀出的內容會根據標籤，利用串列、字典建立具有層次的資料結構，
因此可以容易取出需要的資料內容。
'''
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

import os,sys
fh = open("./data/64_72hr_CH.xml", "r", encoding="utf-8")
xml=fh.read()
fh.close()

!pip install xmltodict
import xmltodict
d = dict(xmltodict.parse(xml))
#累計數量
c = 0
#檢查每個地點
for n in range(0,len(d['cwbopendata']['dataset']['locations']['location'])):
    #是否含有關於溫度的資料
    for m in range(0,len(d['cwbopendata']['dataset']['locations']['location'][n]['weatherElement'])):
        if d['cwbopendata']['dataset']['locations']['location'][n]['weatherElement'][m]['elementName'] == 'T':
            c += 1
            break
print(c)
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
#每個地點
for n in range(0,len(d['cwbopendata']['dataset']['locations']['location'])):
    #尋找關於溫度的資料
    for m in range(0,len(d['cwbopendata']['dataset']['locations']['location'][n]['weatherElement'])):
        if d['cwbopendata']['dataset']['locations']['location'][n]['weatherElement'][m]['elementName'] == 'T':
            print(
                '地點:',d['cwbopendata']['dataset']['locations']['location'][n]['locationName'],'\n'
                '時間:',d['cwbopendata']['dataset']['locations']['location'][n]['weatherElement'][m]['time'][0]['dataTime'],'\n'
                '溫度:',d['cwbopendata']['dataset']['locations']['location'][n]['weatherElement'][m]['time'][0]['elementValue']['value'],
                d['cwbopendata']['dataset']['locations']['location'][n]['weatherElement'][m]['time'][0]['elementValue']['measures'],'\n'
            )
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
print('地點:',d['cwbopendata']['dataset']['locations']['location'][0]['locationName'],'\n')
#尋找關於溫度的資料
for n in range(0,len(d['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'])):
    if d['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][n]['elementName'] == 'T':
        for m in range(0,len(d['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][n]['time'])):
            print(
                '時間:',d['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][n]['time'][m]['dataTime'],'\n'
                '溫度:',d['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][n]['time'][m]['elementValue']['value'],
                d['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][n]['time'][m]['elementValue']['measures'],'\n'
            )
```
