## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests, json
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
data = json.loads(response)
l=[]
print(type(data),'\n')
print('回傳的資料筆數:',len(data),'筆\n')
for d in data:
    for n in d.keys():
        if not n in l:
            l.append(n)
            print(len(l),'',n)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for d in data:
    print('\n標題:',d['title'],
          '\n貼文時間:',d['createdAt'],
          '\n留言人數:',d['commentCount'],
          '\n按讚人數:',d['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
#以like達到100為熱門標準
hot_count=0
not_hot_count=0
hot_commencount=0
hot_likecount=0
not_hot_commencount=0
not_hot_likecount=0

for d in data:
    if d['likeCount'] < 100:
        not_hot_count += 1
        not_hot_commencount += d['commentCount']
        not_hot_likecount += d['likeCount']
    else:
        hot_count += 1
        hot_commencount += d['commentCount']
        hot_likecount += d['likeCount']
    
print('熱門平均留言人數:%d\n熱門平均按讚人數:%d\n非熱門平均留言人數:%d\n非熱門平均按讚人數:%d'
      % (hot_commencount/hot_count,hot_likecount/hot_count,
        not_hot_commencount/not_hot_count,not_hot_likecount/not_hot_count)
     )
```
